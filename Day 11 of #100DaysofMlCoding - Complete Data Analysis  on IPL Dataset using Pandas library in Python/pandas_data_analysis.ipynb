### IPL EDA

```python
# You have to find out top five batsmans in terms of runs scored in all the seasons of IPL!
delivery = pd.read_csv('deliveries.csv')
delivery.head(10)
```

```python
runs = delivery.groupby('batsman')
```

```python
runs.get_group('V Kohli')
```

```python
runs.get_group('V Kohli').shape
```

```python


runs['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
# Find the batsman who hitted largest number of fours in all the seasons.
mask =  delivery['batsman_runs']==6
new_delivery = delivery[mask]
new_delivery.head()
```

```python
max_four = new_delivery.groupby('batsman')
```

```python
max_four['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
# I want to find againest which team virat kohli has scored the highest scores.
mask =  delivery['batsman']== 'V Kohli'
only_kohli = delivery[mask]
only_kohli
```

```python
team = only_kohli.groupby('bowling_team')
```

```python
team['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
# Lets create a function in which you pass name of a batsman as a string it will return you against which team he scored the highest runs. 
def run_scored(batsman_name):
    mask =  delivery['batsman']== batsman_name
    only_kohli = delivery[mask]
    team = only_kohli.groupby('bowling_team')
    return team['batsman_runs'].sum().sort_values(ascending=False).head(3).index[0] 
     
```

```python
run_scored('V Kohli')
```

## isin Function

```python
# Find the most destructive death over batsman in the history of IPL?
#Strike rate =  (number of runs scored/Number of balls)/100 (How many runs batsman scored after playing 100 balls)
# We consider another condition as well i.e., the batsman must have minimum played 200 balls in the 16-20th over.
```

```python
# First of all we will find the number of balls each batsman has played in IPL.
death_over = delivery[delivery['over']>15]
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
X = all_batsman>200
all_batsman[X]
```

```python
all_batsman[X].shape
```

```python
all_batsman[X].index
```

```python
batsman_list = all_batsman[X].index.to_list()
```

```python
# I have to compute run scored by these 43 batsmans.
# I have to count the balls played by these 43 batsmans.
delivery['batsman'].isin(batsman_list)
delivery[delivery['batsman'].isin(batsman_list)]
```

```python
(delivery[delivery['batsman'].isin(batsman_list)]).shape
```

```python
final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```

### Merge Function

```python
# I want to solve a problem and the problem is that I have to find the orange cap holder guy.
# Orange Cap holder is a person who scores the highest runs in a particular season. 
match = pd.read_csv('matches.csv')
new = delivery.merge(match, left_on='match_id', right_on='id')
```

```python
delivery.merge(match, left_on='match_id', right_on='id').shape
```

```python
print(delivery.shape)
print(match.shape)
```

```python
# Multi-index series
new.groupby(['season', 'batsman'])['batsman_runs'].sum()
```

```python
type(new.groupby(['season', 'batsman'])['batsman_runs'].sum())
```

```python
#Convert series to dataframe (reset_index())
new.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset = 'season', keep='first').sort_values('season')[['season','batsman']]
```
