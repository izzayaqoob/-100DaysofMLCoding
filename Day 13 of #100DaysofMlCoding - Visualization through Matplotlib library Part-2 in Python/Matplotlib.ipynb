# **Scatter Plot**

## Scatter plot is used for bi-variate(2 variables) analysis

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1] # x-variable
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1] # y-variable
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5] # values for the opacity of color with varying satisfaction level


sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

#plt.scatter(x, y, s=sizes, c = colors, cmap='Greens', edgecolors='black', linewidths=1, alpha=0.75) # cmap is for the color to be used

plt.scatter(x, y, s=sizes, c = colors, cmap='gist_heat', edgecolors='black', linewidths=1, alpha=0.75)
cbar =plt.colorbar()  # it is for the bar attached for showing the extent of satisfaction level
cbar.set_label('Satisfication Level')
plt.tight_layout() # used for avoiding conjestion in the graph
plt.show()
```

# **Plotting Time Series Data**

```python
from datetime import datetime, timedelta

from matplotlib import dates as mpl_dates # modeule for formatting dates
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle = 'solid',color='red')

plt.gcf().autofmt_xdate() # used for tilting x-axis values for better reading

date_format = mpl_dates.DateFormatter('%d %b, %Y')

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()

plt.show()
```
